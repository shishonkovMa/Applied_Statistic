Most adult mammals are unable to digest lactose contained in milk. In humans, an enzyme called lactase can break down lactose and complete milk digestion. It is encoded by a gene LCT. If that gene has a variant 13910T, the carriers are able to digest milk in adulthood. The distribution of this genetic variant is very different across subpopulations of humans. Out of 50 tested Maya peoples, 13910T was found in one. Build a 95% Wald's confidence interval for the proportion of 13910T carriers in Maya population. What is its' upper bound? Round the answer to 3 decimal points.

```python
import numpy as np
import scipy as sc
from matplotlib import pyplot as plt
from statsmodels.stats.proportion import proportion_confint, proportions_ztest, \
confint_proportions_2indep, test_proportions_2indep


proportion_confint(1, 50, method = 'normal')
```
